```python
print("[ GET ] get scrape date ...")
update_date = ts._get_scrape_date()
print("[ GET ] get talk posted date ...")
talk_date = ts.get_talk_posted_date(ta_soup)
print("[ GET ] get talk titles ...")
talk_titles = ts.get_talk_titles(ta_soup)
print("[ GET ] get talk links ...")
talk_links = ts.get_talk_links(ta_soup)
```

```python
talk_topics = []
talk_transcript = []
transcript_time = []

talk_num = len(talk_links)
ts.all_page_list = talk_num
```

```python
print("[ GET ] get talk topics and transcripts ...")
for i, tl in enumerate(talk_links[:3]):
    ts.target_page_num = i + 1
    ts.target_url = tl
    print("  [{}/{}] Target URL: {}".format(i + 1, talk_num, tl))
    ta_soup = TEDScraper.make_soup(tl)

    # print("          [ GET ] get talk topics")
    # talk_topics.append(ts.get_talk_topics(ta_soup))

    print("          [ GET ] get all language talk transcript")
    talk_transcript.append(ts.get_talk_transcrpit(ta_soup))

    tr_url = TEDScraper.get_transcript_url(tl)
    ts.target_url = tr_url
    print("          Target transcript URL: {}".format(tr_url))
    tr_soup = TEDScraper.make_soup(tr_url)

    print("            [ GET ] get transcript time")
    transcript_time.append(ts.get_talk_transcript_time(tr_soup))
```
